```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import torch

# Verifique se você tem a GPU disponível
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Carregue o modelo YOLO
model5 = YOLO('yolov5n.pt')  # Substitua pelo caminho para o seu modelo, se necessário

# Caminho para a imagem de entrada
image_path = '/content/gettyimages-523602632-612x612.jpg' # Atualize para o caminho correto no Colab

# Realizar a predição
results = model5(image_path)

# Obtenha os bounding boxes e as classes
detections = results[0].boxes  # Primeiro resultado
boxes = detections.xyxy.cpu().numpy()  # Bounding boxes [x_min, y_min, x_max, y_max]
scores = detections.conf.cpu().numpy()  # Confidences
labels = detections.cls.cpu().numpy()  # Class labels

# Carregar a imagem original
image = Image.open(image_path).convert("RGB")
draw = ImageDraw.Draw(image)

# Desenhar os bounding boxes sobre a imagem
for i, box in enumerate(boxes):
    x_min, y_min, x_max, y_max = box
    label = int(labels[i])
    score = scores[i]
    draw.rectangle([x_min, y_min, x_max, y_max], outline="red", width=3)
    draw.text((x_min, y_min - 10), f"Class: {label}, Conf: {score:.2f}", fill="yellow")

# Exibir a imagem com os bounding boxes
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis("off")
plt.show()
```
